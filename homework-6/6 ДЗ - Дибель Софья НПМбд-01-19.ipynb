6.  Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def create_m(n):
    matrix = np.zeros((n, n))
    for i in range(n):
        matrix[i, :] = np.arange(0, n) + i
    return matrix

print(create_m(10))
```
