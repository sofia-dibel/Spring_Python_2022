```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import scipy
```

## Exercise 1  

### 1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!

```python
xx = np.linspace(-1, 1, 250)
yy = np.power(xx, 2) * np.sin(1 / np.power(xx, 2)) + xx
fig, ax = plt.subplots(figsize=(8, 8), facecolor='#FFFFE0')
ax.plot(xx, yy, label='y=y(x)', color='#FFD700', linewidth=2)
plt.xlim(-1, 1)
plt.ylim(-0.75, 1.25)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('$x^2⋅sin(1/x^2)+x$')
ax.legend()
```

### 2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
xx = np.linspace(5, 25, 250)
yy = abs(1 - np.power(xx, 2) / (1 + np.power(xx, 2)))
fig, ax = plt.subplots(figsize=(7, 7))
ax.semilogy(xx, yy, linestyle='--', color='#4B0082',  label='y=y(x)', linewidth='3', alpha=0.66)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('разница функций $1/(1+x^2)$ и $1/x^2$ ')
ax.legend()
```

## Гамма-фукнция: $\Gamma \left( x \right) = \int\limits_0^\infty {s^{x - 1} e^{ - s} ds}$

```python
xx = np.linspace(-3.5, 5.5, 2251)
yy = scipy.special.gamma(xx)
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(xx, yy, color='#6B8E23', alpha=0.8, linewidth=3, label='y=y(x)')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
plt.grid()
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('Гамма-фукцния')
ax.legend(bbox_to_anchor=(1.02, 1))
```

## Факториал y = x!

```python
xx = np.linspace(0, 5, 6)
yy = scipy.special.factorial(xx, exact=False)
fig, ax = plt.subplots(figsize=(7, 7))
ax.bar(xx, yy, color='#7B68EE', alpha=0.5, linewidth=4, label='y=x!', linestyle='--')
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('Факториал')
ax.legend()
ax.legend(bbox_to_anchor=(1.02, 1))
```

## Exercise 2  
### 1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
xx = np.linspace(- 6 * np.pi, 6 * np.pi, 300)
yy = np.sin(xx) / xx
fig, ax = plt.subplots(figsize=(15, 6))

ticks = np.linspace(- 6 * np.pi, 6 * np.pi, 25)
tickslabels = np.array([])
i = -6
while i <= 6:
    if i == 0:
        tickslabels = np.append(tickslabels, '0')
    elif i == -1:
        tickslabels = np.append(tickslabels, '$-\pi$')
    elif i == 1:
        tickslabels = np.append(tickslabels, '$\pi$')
    elif i == -0.5:
        tickslabels = np.append(tickslabels, '$-\pi/2$')
    elif i == 0.5:
        tickslabels = np.append(tickslabels, '$\pi/2$')
    elif i % 1 == 0:
        tickslabels = np.append(tickslabels, str(int(i)) + '$\pi$')
    elif i % 1 == 0.5:
        tickslabels = np.append(tickslabels, str(int(2 * i)) + '/2$\pi$')
    i += 0.5
        
ax.set_xticks(ticks)
ax.set_xticklabels(tickslabels)
x_minor_ticker = matplotlib.ticker.FixedLocator(np.linspace(- 11 * np.pi / 2, 11 * np.pi / 2, 12))
ax.xaxis.set_minor_locator(x_minor_ticker)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_title('$sin(x)/x$')
ax.grid()
ax.plot(xx, yy, label='y=y(x)', color='#006400', alpha=0.7, linewidth=3, linestyle='-')
ax.legend()
```

## Exercise 3  
### 1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.

```python
xx = yy = np.linspace(-np.pi, np.pi, 300)
X, Y = np.meshgrid(xx, yy)
Z = np.sin(X) * np.cos(Y)
fig, ax = plt.subplots(figsize=(10, 8))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm=norm, cmap='cubehelix', shading='auto')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)

cb = fig.colorbar(p, ax=ax)
cb.set_label('z', fontsize=15)
```

### 2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
xx = yy = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(xx, yy)
Z = np.exp(- np.power(X, 2) - np.power(Y, 2))
fig, ax = plt.subplots(figsize=(10, 8))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm=norm, cmap='viridis', shading='auto')


ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)

cb = fig.colorbar(p, ax=ax)
cb.set_label('z', fontsize=15)
```
